# Pivoter les données


Il s'agit ici de réorganiser les données d'une forme longue en une forme large ou de transformer les lignes en colonnes ou vice versa, exactement comme la fonction pivoter les données dans Excel.

```python
# Créons un exemple de dataframe

df = pd.DataFrame({'Mois': ["Janvier", "Janvier", "Janvier", "Janvier",
                                  "Fevrier", "Fevrier", "Fevrier", "Fevrier",
                                  "Mars", "Mars", "Mars", "Mars"],
                   'Categorie': ["Transport", "Courses", "Maison", "plaisir",
                                "Transport", "Courses", "Maison", "Plaisir",
                                "Transport", "Courses", "Maison", "Plaisir"],
                   'Montant': [78., 215., 175., 120., 115., 240., 325., 125., 90., 260., 270., 120.]})
```

```python
# affichons le type de df
type(df)
```

```python
df
```

```python
# nous pouvons faire la somme des colonnes numériques

df['Montant'].sum(axis = 0)
```

```python
df
```

```python
# pivotons notre dataframe
df_pivot = df.pivot(index="Mois", columns="Categorie", values="Montant")
df_pivot
```

```python
#Nous pouvons calculer les sommes des valeurs par ligne ou par colonne
# la valeur de l'axe des lignes est 0 et celles des colonne est 1

#sommes par lignes
df_pivot.sum(axis=0)
```

```python
# sommes par colonnes
df_pivot.sum(axis=1)
```

# Nous pouvons aggreger les données en pivotant

```python
df = pd.read_csv("titanic.csv")
```

```python
# ici nous aggregeons le données en lignes par genre et en colonne par classe des passagers
# les valeur des tickets sont également aggrégées selons les combinaisons de ligne et de colonnes
df.pivot_table(index='Sex', columns='Pclass', values='Fare')
```

<div class="alert alert-info">

<b>RAPPEL</b>:

 <ul>
  <li>Par defaut, "pivot_table" prend la  <b>moyenne</b> de toutes les valeurs regroupées dans une cellule. Vous pouvez toutefois spécifier la fonction d'aggrégation que vous voulez utiliser avec les mot clé <b>aggfunc</b>.</li>
</ul>
</div>

```python
# Exemple : pivoter en prenant la valeur maximale des prix des tickets pour chaque cellule (max ici est AVERAGE).

df.pivot_table(index='Sex', columns='Pclass',
               values='Fare', aggfunc='max')
```

```python
# pivoter en prenant la valeur minimale des prix des tickets pour chaque cellule
df.pivot_table(index='Sex', columns='Pclass',
               values='Fare', aggfunc='min')
```

```python
# pivoter en comptant le nombre des prix des tickets pour chaque cellule
df.pivot_table(index='Sex', columns='Pclass',
               values='Fare', aggfunc='count')
```

<div class="alert alert-info">

<b>RAPPEL</b>:

 <ul>
  <li>il y a une fonction raccourci pour <b>pivot_table</b> avec `aggfunc=count` comme fonction c'est : <b>crosstab</b>
  </li>
</ul>
</div>

```python
# exemple
pd.crosstab(index=df['Sex'], columns=df['Pclass'])
```

# La fonction Melt
C'est l'inverse de la fonction pivoter, il permet d'allonger une table de données compacte.

```python
# exemple créons un dataframe qui montre la moyenne des prix des tickets par classe et par sexe
# nous ne voulons pas de nom de colonnes
pivoted = df.pivot_table(index='Sex', columns='Pclass', values='Fare').reset_index()
pivoted.columns.name = None
```

```python
#notre dataset se présente ainsi
pivoted
```


Pour avoir toutes les observations dans une seule colonne, nous pouvons utiliser la fonctions <b>melt</b>

```python
pd.melt(pivoted)
```


Même si la transformation a été faite avec toutes les catégories dans une colonne et les valeurs dans une autre colonne, ce n'est pas exactement ce que nous voulons.Il serait préférable d'avoir la colonne <b>sex</b> séparée.

```python
# pour se faire nous allons inclure des arguments dans la fonction
pd.melt(pivoted, id_vars=['Sex']) #, var_name='Pclass', value_name='Fare')
```

# Modifier la forme des données avec `stack` et `unstack`

Pour utiliser `stack`/`unstack`, nous avons besoin que la valeur que nous voulons changer des lignes en colonnes ou dans l'autre sens soit l'index:

```python
# Exemple de données

df = pd.DataFrame({'A':['one', 'one', 'two', 'two'],
                   'B':['a', 'b', 'a', 'b'],
                   'C':range(4)})
df
```

```python
df = df.set_index(['A', 'B']) # nous pouvons combiner dex index
df
```

```python
result = df['C'].unstack()
result
```

```python
df = result.stack().reset_index(name='C')
df
```

<div class="alert alert-info">

<b>RAPPEL</b>:

 <ul>
  <li>**stack**: rend les données plus  <b>longues</b> et <b>compactes</b> </li>
  <li>**unstack**: rend les données plus <b>courtes</b> et  <b>larges</b> </li>
</ul>
</div>
